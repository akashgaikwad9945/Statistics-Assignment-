```python
import pandas as pd
import numpy as np
import seaborn as sns
import statistics as st
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

```python
int_list2 = np.random.randint(200,300,500)
int_list2 = int_list2.tolist()
```

```python
plt.figure(figsize=(10, 6))
```

```python
# 1. Frequency & Gaussian Distribution
plt.figure(figsize=(10, 6))  # Set figure size

# Frequency plot
plt.hist(int_list2, bins='auto', alpha=0.7, label='Frequency')


mu = np.mean(int_list2) 
sigma = np.std(int_list2)  
x = np.linspace(min(int_list2), max(int_list2), 100)  # Range for the Gaussian curve
plt.plot(x, norm.pdf(x, mu, sigma), label='Gaussian Distribution')

plt.xlabel('Data Values')
plt.ylabel('Frequency/Probability Density')
plt.title('Frequency & Gaussian Distribution')
plt.legend()
plt.grid(True)
plt.show()

```

```python
def calculate_range(int_list2):
  """
  This function calculates the range of a list of numbers.

  Args:
      int_list2: A list of numerical values.

  Returns:
      The range of the data (difference between the maximum and minimum values), 
      or None if the list is empty.
  """

  if not int_list2:
    return None  # Handle empty list

  # Find the minimum and maximum values
  min_value = min(int_list2)
  max_value = max(int_list2)

  # Calculate the range
  range_value = max_value - min_value

  return range_value
```

```python
def calculate_iqr(int_list2):
  """
  This function calculates the interquartile range (IQR) of a list of values.

  Args:
      int_list2: A list of numerical values.

  Returns:
      The interquartile range (IQR) of the data, or None if the list is empty or has less than 4 elements.
  """

  if not int_list2 or len(int_list2) < 4:
    return None  # Handle empty list or list with less than 4 elements

  # Sort the data
  sorted_data = sorted(int_list2)

  # Calculate quartiles (assuming even number of elements)
  q1_index = int(len(sorted_data) / 4) - 1
  q3_index = int(3 * len(sorted_data) / 4) - 1

  # Handle odd number of elements (take medians of middle two elements)
  if len(sorted_data) % 2 != 0:
    q1_index = int(len(sorted_data) / 2) - 2
    q3_index = int(len(sorted_data) / 2) + 1

  # Calculate quartiles
  q1 = sorted_data[q1_index]
  q3 = sorted_data[q3_index]

  # Calculate IQR
  iqr = q3 - q1

  return iqr
```

```python
def calculate_cv(int_list2):
  """
  This function calculates the coefficient of variation (CV) for a dataset.

  Args:
      int_list2: A list of numerical values.

  Returns:
      The coefficient of variation (CV) as a float, or None if the list is empty or has a standard deviation of zero.

  Raises:
      ValueError: If the standard deviation is zero, resulting in division by zero.
  """

  if not int_list2:
    return None  # Handle empty list

  # Calculate standard deviation (assuming you've imported NumPy as np)
  std_dev = np.std(int_list2)

  # Check for zero standard deviation (division by zero error)
  if std_dev == 0:
    raise ValueError("Standard deviation is zero. Coefficient of variation cannot be calculated.")

  # Calculate mean
  mean_value = np.mean(int_list2)

  # Calculate coefficient of variation
  cv = std_dev / mean_value

  return cv
```

```python
def calculate_mad(int_list2):
  """
  This function calculates the mean absolute deviation (MAD) of a list of numbers.

  Args:
      int-LIST2: A list of numerical values.

  Returns:
      The mean absolute deviation (MAD) as a float, or None if the list is empty.
  """

  if not int_list2:
    return None  # Handle empty list

  # Calculate the mean
  mean_value = np.mean(int_list2)

  # Calculate absolute deviations from the mean
  absolute_deviations = [abs(x - mean_value) for x in int_list2]

  # Calculate MAD (average of absolute deviations)
  mad = np.mean(absolute_deviations)

  return mad
```

```python
def calculate_quartile_deviation(int_list2):
  """
  This function calculates the quartile deviation (QD) of a list of values.

  Args:
    int_list2: A list of numerical values.

  Returns:
      The quartile deviation (QD) as a float, or None if the list is empty or has less than 4 elements.
  """

  if not int_list2 or len(int_list2) < 4:
    return None  # Handle empty list or list with less than 4 elements

  # Sort the data
  sorted_data = sorted(int_list2)

  # Calculate quartiles (assuming even number of elements)
  q1_index = int(len(sorted_data) / 4) - 1
  q3_index = int(3 * len(sorted_data) / 4) - 1

  # Handle odd number of elements (take medians of middle two elements)
  if len(sorted_data) % 2 != 0:
    q1_index = int(len(sorted_data) / 2) - 2
    q3_index = int(len(sorted_data) / 2) + 1

  # Calculate quartiles
  q1 = sorted_data[q1_index]
  q3 = sorted_data[q3_index]

  # Calculate quartile deviation
  quartile_deviation = (q3 - q1) / 2

  return quartile_deviation

```

```python
def calculate_range_based_coefficient(int_list2):
  """
  This function calculates the range-based coefficient of dispersion for a dataset.

  Args:
     int_list2: A list of numerical values.

  Returns:
      The range-based coefficient of dispersion as a float, or None if the list is empty.
  """

  if not int_list2:
    return None  # Handle empty list

  # Find minimum and maximum values
  min_value = min(int_list2)
  max_value = max(int_list2)

    
  # Calculate range
  range_value = max_value - min_value

  # Calculate coefficient of dispersion
  if range_value == 0:
    # Handle case where all elements are equal (coefficient is undefined)
    return None
  else:
    coefficient = (max_value - min_value) / (max_value + min_value)

  return coefficient
```
